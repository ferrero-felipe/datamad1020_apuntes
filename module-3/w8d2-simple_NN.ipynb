```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import numpy as np
# np.random.seed(2)
```

```python
# First of all, our data!
X = np.array([0.1,0.15,0.2,0.45,0.5,0.55,0.9,0.95,1])
y = np.array([0,0,0,1,1,1,0,0,0])
```

```python
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage");
```

```python
# In reality, that is not enough!
X = np.concatenate([np.linspace(0,0.25,1000), np.linspace(0.4,0.6,1000), np.linspace(0.8,1,1000)])
y = np.array([0]*1000+[1]*1000+[0]*1000)

# That's more like it.
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage");
```

```python
# Some train_test_splitting
# X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
# Now, our NN!

# The keras base for sequential models
model = Sequential()

# The hidden layer
# We input the numer of nodes, the input size as this the first layer and the activation function
model.add(Dense(2,input_dim=1,activation="softplus"))

# Then the output layer
# Since we will not put an activation function, it shall be a simple sum.
# We must do this to get back to 1 node
model.add(Dense(1))
```

```python
## Some info
# Keras dense layer
# https://keras.io/api/layers/core_layers/dense/
# https://www.tutorialspoint.com/keras/keras_dense_layer.htm

# Activation functions
# https://keras.io/api/layers/activations/
```

```python
# Now we compile!
# We just must set the loss function and the optimizer.

opt = Adam(learning_rate=0.1)
model.compile(loss="mean_squared_error", optimizer=opt)
```

```python
# And we train!
# We must define the number of epochs ( How many times we will go through the data (and backpropagate))
history = model.fit(X, y, epochs=500, verbose=2)
```

```python
# And we can print our green squigle.
x = np.linspace(0,1,1000)
x = np.expand_dims(x,-1)
y_pred = model.predict(x)
```

```python
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage")
plt.plot(x,y_pred, c="green");
```

```python
plt.grid()
plt.plot(history.history["loss"])
plt.ylabel("Loss")
plt.xlabel("Epochs");
```

## BAM!

```python
# We can even check the weights
conn = []
for layer in model.layers:
    weight = layer.get_weights()
    conn.append(weight)
```

```python
conn1 = lambda x: -17.4*x -6.7
conn2 = lambda x: 8.3*x +2.8
```

```python
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage")
plt.plot(x,conn1(x), label="connection 1")
plt.plot(x,conn2(x), label="connection 2")
plt.legend();
```

$f(x)= ln(1+e^x)$

```python
soft_plus = lambda x: np.log(1+np.exp(x))
```

```python
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage")
plt.plot(x,soft_plus(conn1(x)), label="Node 1")
plt.plot(x,soft_plus(conn2(x)), label="Node 2")
plt.legend();
```

```python
plt.plot(x,soft_plus(conn1(x)), label="Node 1")
```

```python
plt.plot(x,soft_plus(conn2(x)), label="Node 2")
```

```python
conn3 = lambda x: -1.4*x 
conn4 = lambda x: 4.3*x 
```

```python
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage")
plt.plot(x,conn3(soft_plus(conn1(x))), label="Connection 3")
plt.plot(x,conn4(soft_plus(conn2(x))), label="Connection 4")
plt.legend();
```

```python
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage")
plt.plot(x,conn3(soft_plus(conn1(x)))+conn4(soft_plus(conn2(x))) -0.2, label="Output")
plt.legend();
```

```python
# Now with ReLU!
model = Sequential()
model.add(Dense(2,input_dim=1))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(1))
model.add(LeakyReLU(alpha=0.1))
```

### Why Leaky ReLU???
[Dying ReLU](https://datascience.stackexchange.com/questions/5706/what-is-the-dying-relu-problem-in-neural-networks)

Its  is still finicky however.

```python
opt = Adam(learning_rate=0.01)
model.compile(loss="mean_squared_error", optimizer=opt)
```

```python
X,y = shuffle(X,y)
```

```python
history = model.fit(X, y, epochs=100,verbose=0)
```

```python
x = np.linspace(0,1,1000)
x = np.expand_dims(x,-1)
y_pred = model.predict(x)
```

```python
plt.grid()
plt.scatter(X,y)
plt.ylabel("Efficiency")
plt.xlabel("Dosage")
plt.plot(x,y_pred, c="green");
```
